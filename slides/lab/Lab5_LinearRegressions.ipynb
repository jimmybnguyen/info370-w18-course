```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import statsmodels.formula.api as smf
import seaborn.apionly as sns
```

```python
iris = pd.DataFrame(data = sns.load_dataset('iris'))
iris
```

```python
# StatsModels linear model
slm = smf.ols('sepal_length ~ petal_length', data = iris).fit()
print(slm.summary())
```

```python
# StatsModels linear model plotting
iris['pred'] = pd.Series(slm.predict())
plt.scatter(iris.petal_length, iris.sepal_length)
plt.plot(iris.petal_length, iris.pred)
plt.show()
```

```python
# SKLearn linear model
regr = linear_model.LinearRegression()
skl = regr.fit(iris[['petal_length']], iris['sepal_length'])
print(skl)
print(skl.intercept_, skl.coef_)
```

```python
# SKLearn linear model plotting
skl.pred = regr.predict(iris[['petal_length']])
plt.scatter(iris.petal_length, iris.sepal_length)
plt.plot(iris.petal_length, skl.pred)
plt.show()
```
